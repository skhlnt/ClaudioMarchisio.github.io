```python
# actually doing it 
import markdown
import os
all_posts = os.listdir('../_posts')
all_md_posts = [post for post in all_posts if post[-2:]=='md']
for postname in all_md_posts:
    with open('../_posts/'+postname, 'r', encoding='UTF-8') as f:
        text = f.read()
    date = text.split('\ndate: ')[1].split(' ')[0]
    title = text.split('\ntitle: ')[1].split('\n')[0]
    formatted_postname = date + '-' + title + '.md'
    month = formatted_postname.split('-')[1]
    day = formatted_postname.split('-')[2]
    if(len(month)<2): month = '0' + month
    if(len(day)<2): day = '0' + day
    formatted_postname = '-'.join([formatted_postname.split('-')[0]]+[month]+[day]+formatted_postname.split('-')[3:])
    print(formatted_postname)
    os.rename('../_posts/'+postname, '../_posts/'+formatted_postname)
```
